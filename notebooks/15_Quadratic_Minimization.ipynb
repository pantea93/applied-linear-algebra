**Quadratic Minimization:**

Minimization of quadratic functions is essential in optimization and least squares problems.

General Form:
  \begin{equation*}
    p(x) = ax^2 + 2bx + c
  \end{equation*}
Cases: 

- $a > 0$: Parabola opens upward, unique minimum.
- $a < 0$: No minimum, parabola opens downward.
- $a = 0$: Linear function, no minimum/maximum unless constant.


**Finding the Minimum**

**Using Calculus:**

  \begin{equation*}
    p'(x) = 2ax + 2b = 0 \Rightarrow x^* = -\frac{b}{a}, \quad p(x^*) = c - \frac{b^2}{a}
  \end{equation*}
  
 Second derivative test:
  \begin{equation*}
    p''(x) = 2a > 0 \Rightarrow x^* \text{ is a minimum.}
  \end{equation*}

**Multi-Variable Case**

Quadratic Function:

  \begin{equation*}
    p(x) = x^T K x - 2x^T f + c, \quad x \in \mathbb{R}^n
  \end{equation*}
  
Matrix Form:

- $K$: Symmetric $n \times n$ matrix.
- $f$: Constant vector.
- $c$: Constant scalar.

**Theorem: Minimization Criterion**

Theorem: If $K$ is positive definite, then $p(x)$ has a unique minimizer:
  \begin{equation*}
    x^* = K^{-1} f
  \end{equation*}
  Minimum value:
  \begin{equation*}
    p(x^*) = c - f^T K^{-1} f
  \end{equation*}


**Example**

  Given:
  \begin{equation*}
    p(x_1, x_2) = 4x_1^2 - 2x_1x_2 + 3x_2^2 + 3x_1 - 2x_2 + 1
  \end{equation*}
  
Matrix Form:
  \begin{equation*}
    K = \begin{bmatrix} 4 & -1 \\ -1 & 3 \end{bmatrix}, \quad f = \begin{bmatrix} -3/2 \\ 1 \end{bmatrix}, \quad c = 1
  \end{equation*}
  
Solving $Kx = f$ yields:
  \begin{equation*}
    x^* = \begin{bmatrix} -7/22 \\ 5/22 \end{bmatrix}, \quad p(x^*) = 13/44
  \end{equation*}

```python
import numpy as np
import matplotlib.pyplot as plt

# ---- Step 1: Define coefficients ----
# p(x, y) = 5x^2 + 3xy + 2y^2 - 4x + 6y - 7
K = np.array([[5, 3/2],
              [3/2, 2]])   # symmetric matrix
f = np.array([2, -3])       # vector from linear terms
c = -7                      # constant term

# ---- Step 2: Solve Kx = f to find minimizer ----
x_star = np.linalg.solve(K, f)
print("Minimizer (x*, y*):", x_star)

# ---- Step 3: Compute minimum value p(x*) = c - f^T K^{-1} f ----
p_min = c - f.T @ np.linalg.inv(K) @ f
print("Minimum value p(x*):", p_min)

# ---- Step 4: (Optional) visualize the quadratic surface ----
x_vals = np.linspace(-4, 4, 100)
y_vals = np.linspace(-4, 4, 100)
X, Y = np.meshgrid(x_vals, y_vals)

# Compute p(x, y) on the grid
P = 5*X**2 + 3*X*Y + 2*Y**2 - 4*X + 6*Y - 7

# ---- Step 5: Plot the surface ----
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, P, cmap='viridis', alpha=0.8)
ax.scatter(x_star[0], x_star[1], p_min, color='r', s=60, label='Minimum Point')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('p(x, y)')
ax.set_title('Quadratic Function Surface')
ax.legend()
plt.show()
```

**Conclusion**

- Quadratic minimization is fundamental in optimization.
- Positive definite matrices ensure a unique minimizer.
- Matrix techniques provide efficient solutions.
